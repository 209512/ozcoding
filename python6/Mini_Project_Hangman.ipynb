#### 과제 설명

이번 개발 주제는 '행맨' 게임이구요.

행맨 게임은 아주 유명한 단어 퀴즈 프로그램이예요

어떤 단어가 주어지면 그 단어의 길이만큼 빈 칸(밑줄) 이 주어지고
플레이어는 알파벳을 하나씩 얘기하면서 단어를 추측해요!

7번 틀리게 되면 게임이 끝납니다. 🥲



#### 예시를 들어볼께요!!

만약 E 가 단어에 포함돼있다면, 해당하는 위치에 E 를 표시해줍니다

_ _ _ _ E E
이렇게 되겠네요.

이번에는 A를 해볼께요
A 는 포함되지 않으므로 기회가 줄어들고 지금까지 맞힌 내용 그대로 출력합니다.

_ _ _ _ E E
변화가 없지요.

이번에는 C 를 해봅시다.

그럼
C _ _ _ E E

다음에는 F 를
C _ F F E E

마지막으로 O 를 해보면
C O F F E E 로 게임은 성공으로 끝나게 됩니다.

```python
import random

words = ['apple', 'banana', 'orange', 'grape', 'lemon']
hangman_pics = [
    """
     ------
     |    |
     |
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |    |
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   /
     |
    ---""",
    """
     ------
     |    |
     |    O
     |   /|\\
     |   / \\
     |
    ---""",
]
    
class HangmanGame:
    def __init__(self):
        self.word = random.choice(words)
        self.guesses = set()
        self.attempts = 6

    def display(self):
        result =''.join([char if char in self.guesses else '_' for char in self.word])
        return result
    
    def get_guess(self):
        while True:
            guess = input('글자를 맞춰보세요: ').lower()
            if len(guess) != 1 or not guess.isalpha():
                print('알파벳 한 개를 입력해주세요.')
            elif guess in self.guesses:
                print('이미 추측한 글자입니다. 다른 글자를 시도하세요.')
            else:
                return guess

    def play(self):
        print('행맨 게임에 오신 것을 환영합니다!')
        while self.attempts > 0:
            print(hangman_pics[6 - self.attempts])
            print('단어:', self.display())
            print('-----------------------')
            guess = self.get_guess()
            
            if guess in self.word:
                self.guesses.add(guess)
                print('단어를 맞추셨군요. 더 힘내봐요.')
                if all(letter in self.guesses for letter in self.word):
                    print(f'축하합니다! 단어를 맞췄습니다. 정답은 {self.word}!')
                    break
            else:
                self.attempts -= 1
                print(f'틀렸습니다. 남은 시도 횟수: {self.attempts}번')
                if self.attempts == 0:
                    print(hangman_pics[-1])
                    print(f'게임 오버! 정답은 {self.word}입니다.')     
                
if __name__ == "__main__":
    game = HangmanGame()
    game.play()
```
